```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("RTA Dataset.csv.zip")

# Explore the dataset
print(df.head())

# Analyze patterns related to road conditions, weather, and time of day
# Countplot for Road Conditions
plt.figure(figsize=(10, 6))
sns.countplot(x='Day_of_week', data=df, palette='viridis')
plt.title('Accidents by Day_of_week')
plt.xticks(rotation=45)
plt.show()

# Countplot for Weather Conditions
plt.figure(figsize=(10, 6))
sns.countplot(x='Age_band_of_driver', data=df, palette='viridis')
plt.title('Age_band_of_driver')
plt.xticks(rotation=45)
plt.show()

# Countplot for Time of Day
plt.figure(figsize=(10, 6))
sns.countplot(x='Sex_of_driver', data=df, palette='viridis')
plt.title('Sex_of_driver')
plt.xticks(rotation=45)
plt.show()
```
